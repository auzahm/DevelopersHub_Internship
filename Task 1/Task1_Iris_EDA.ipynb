## Dataset Description
The Iris dataset contains 150 samples of three Iris species (`setosa`, `versicolor`, `virginica`).  
It includes the following features:
- `sepal_length` (cm)
- `sepal_width` (cm)
- `petal_length` (cm)
- `petal_width` (cm)
- `species` (categorical target)


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')

print("Shape:", iris.shape)
print("Columns:", iris.columns)
iris.head()
```

We can see the dataset has 150 rows and 5 columns, including the target column `species`.


```python
iris.isnull().sum()
```

There are no missing values in the dataset, so no cleaning is required. All data is numeric except the `species` column.


```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(iris['petal_length'], bins=20, kde=True)
plt.title("Petal Length Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title("Sepal Length by Species")
plt.show()
```

From the visualizations:
- Setosa is clearly separable in sepal/petal features.
- Petal length and width show the most distinction between species.
- Some outliers exist in sepal width.


## Conclusion
- Iris dataset has 150 samples with 4 numeric features and 3 species classes.
- Visualizations revealed clear separation of species based on petal measurements.
- Random Forest classifier achieved high accuracy (~100%) on this dataset.
